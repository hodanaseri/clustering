```python
#loading libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
# Generate random number between 1500 & 2500 
sample = np.random.randint(1500, 2500)
x = np.random.rand(sample)
y = np.random.rand(sample)
# Create a scatter plot to show the distribution of the samples
plt.scatter(x, y)
plt.title(f'Distribution of {sample} Random Samples')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
data = np.column_stack((x, y))

k_values = [2, 3, 4, 5]

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(data)
    
    # Plot the clustered data
    plt.scatter(x, y, c=labels, s=10, cmap='viridis')
    plt.title(f'K-means Clustering (k={k})')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
from sklearn.datasets import make_blobs

sample = np.random.randint(1500, 2500)
X, _ = make_blobs(n_samples=sample, centers=5, random_state=42)

#different values of K
k_values = [2, 3, 4, 5, 6]

plt.figure(figsize=(12, 10))

for i, k in enumerate(k_values, 1):
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)

    # Visualize the clusters
    plt.subplot(3, 2, i)
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, alpha=0.7)
    plt.scatter(
        kmeans.cluster_centers_[:, 0],
        kmeans.cluster_centers_[:, 1],
        c='red',
        marker='x',
        s=200,
        label='Centroids',
    )
    plt.title(f'K-means Clustering with K={k}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
def samplegen(sample):
    np.random.seed(42)

    cluster_centers = np.array([[1, 1], [1, 4], [4, 1], [4, 4]])
    samples = np.zeros((sample, 2))
    
    for i in range(sample):
        cluster = np.random.randint(4)
        samples[i] = np.random.normal(loc=cluster_centers[cluster], scale=0.5)
    
    return samples

sample = np.random.randint(1500, 2500)
data = samplegen(sample)

plt.scatter(data[:, 0], data[:, 1], s=10, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
k_values = [2, 3, 4, 5, 6]

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(data)
    centroids = kmeans.cluster_centers_

    plt.scatter(data[:, 0], data[:, 1], c=labels, s=10)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
    plt.title(f'K-means Clustering with K={k}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
np.random.seed(42)
sample = np.random.randint(1500, 2500)
X, _ = make_blobs(n_samples=sample, centers=4, cluster_std=0.5, random_state=42)
k_values = range(1, 10)
reconstruction_errors = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    reconstruction_errors.append(kmeans.inertia_)

plt.plot(k_values, reconstruction_errors, marker='o')
plt.xlabel('Clusters')
plt.ylabel('Reconstruction Error (Inertia)')
plt.xticks(k_values)
plt.show()
```

```python
X, _ = make_blobs(n_samples=1000, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

centroid_distances = np.min(kmeans.transform(X), axis=1)
reconstruction_error = np.mean(centroid_distances)
print("Reconstruction error:", reconstruction_error)
```
